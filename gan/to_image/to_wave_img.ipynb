```python
# 디노이즈 한 걸로 보자...^^

# 목소리 불러와~
gan_f_y_d, sr = librosa.load('C:/nmb/gan_0504/audio/b300_e1000_n500/b300_e1000_n500_total00010_sum.wav', sr=22050) 

print(len(gan_f_y_d))
print('Audio length (seconds): %.2f' % (len(gan_f_y_d) / sr))

plt.figure(figsize=(15, 3))
plt.title('2d sound wave graph_gan_F(e10)')
librosa.display.waveplot(y=gan_f_y_d, sr=sr)
plt.ylabel("Amplitude")
```

```python
# 4) Mel Spectogram
S = librosa.feature.melspectrogram(gan_f_y_d, sr=sr, n_fft=512, hop_length=128, center=True)
S_DB = librosa.amplitude_to_db(S, ref=np.max)

plt.figure(figsize=(25, 4))
plt.title('Mel Spectogram_gan_F(e10)')
librosa.display.specshow(S_DB, sr=sr, hop_length=512, x_axis='time', y_axis='log')
plt.colorbar()
plt.show()
```
