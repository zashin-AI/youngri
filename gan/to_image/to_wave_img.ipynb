```python
# gan06 에서 만들어진 음성을..wave 이미지로 바꿔보자

# 1) 2D 음파 그래프

import librosa
import matplotlib.pyplot as plt
import librosa.display
import numpy as np

# 목소리 불러와~
gan_f_y, sr = librosa.load('C:/nmb/gan_0504/audio/b300_e1000_n500/b300_e1000_n500_total1000_sum.wav', sr=22050) 

print(len(gan_f_y))
print('Audio length (seconds): %.2f' % (len(gan_f_y) / sr))
```

```python
plt.figure(figsize=(12, 3))
plt.title('2d sound wave graph_gan_F')
librosa.display.waveplot(y=gan_f_y, sr=sr)
plt.ylabel("Amplitude")
```

```python
# 4) Mel Spectogram
S = librosa.feature.melspectrogram(gan_f_y, sr=sr, n_fft=512, hop_length=128, center=True)
S_DB = librosa.amplitude_to_db(S, ref=np.max)

plt.figure(figsize=(10, 4))
plt.title('Mel Spectogram')
librosa.display.specshow(S_DB, sr=sr, hop_length=512, x_axis='time', y_axis='log')
plt.colorbar()
plt.show()

# 무서워....
```
